```python
import re
import requests
from bs4 import BeautifulSoup
import time
import os
import pandas as pd
import numpy as np
```

```python
os.getcwd()
path = 'yuanEV'
```

```python
data = []

my_headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36', 'Accept-Language':'zh-CN'}
AUTOTHROTTLE_ENABLED = True

page = 97

for k in range(1, page+1):
    if k == 1 :
        url = "https://www.maiche.com/repute/s3966/"
    else :
        url = "https://www.maiche.com/repute/s3966-p" +str(k)+ "/#headNav#fixed"
        
    src = False
    for i in range(1,6):
        try:
            response = requests.get(url, headers = my_headers)
            src = response.content
            break
        except:
            print('failed attempt #', i)
            time.sleep(2)
    if not src:
        print('Could not get page', url)
    else:
        print('Successfully get the page', url)
        
    soup = BeautifulSoup(src.decode('utf-8', 'ignore'),'lxml')
    
    infos = soup.findAll('div', {'class':re.compile('car-group clearfix')})
    for info in infos:
        left = info.findAll('div', {'class':re.compile('gleft')})
        right = info.findAll('div', {'class':re.compile('kb-all active')})
        
        for cus in left:
            cus_1 = cus.findAll('div',{'class':re.compile('head-info')})
            cus_2 = cus.findAll('div',{'class':re.compile('g2')})
            cus_3 = cus.findAll('div',{'class':re.compile('g3')})
            
        for rev in right:
            rev_1 = rev.findAll('div',{'class':re.compile('line')})
            
            for a in cus_1:
                try:
                    cus_name = a.find('div',{'class':re.compile('name')})
                    cus_time = a.find('div',{'class':re.compile('lbl')})
                except:
                    continue
            for b in cus_2:
                try:
                    c_type = b.find('h3')
                    c_info1 = b.findAll('label',{'class':re.compile('value')})
                except:
                    continue
            for c in cus_3:
                try:
                    scores = c.findAll('span',{'class':re.compile('num')})
                except:
                    continue
            for rew in rev_1:
                try:
                    adv = rev_1[0].find('span',{'class':re.compile('desc')})
                    disadv = rev_1[1].find('span',{'class':re.compile('desc')})
                    room = rev_1[2]
                    power = rev_1[3]
                    control = rev_1[4]
                    fuel = rev_1[5]
                    comfort = rev_1[6]
                    outer_view = rev_1[7]
                    inner_view = rev_1[8]
                    price = rev_1[9]
                except:
                    continue
            
            try:
                customer_name = cus_name.text.strip()
                present_time = cus_time.text.strip()
                car_type = c_type.text.strip()
                purchase_time = c_info1[0].text.strip()
                purchase_place = c_info1[1].text.strip()
                car_price = c_info1[2].text.strip()
                average_mpg = c_info1[3].text.strip()
                drive_distance = c_info1[4].text.strip()
                overall_score = scores[0].text.strip()
                room_score = scores[1].text.strip()
                power_score = scores[2].text.strip()
                control_score = scores[3].text.strip()
                mpg_score = scores[4].text.strip()
                comfort_score = scores[5].text.strip()
                outer_score = scores[6].text.strip()
                cost_score = scores[7].text.strip()
                adv_review = adv.text.strip()
                dis_review = disadv.text.strip()
                room_review = str(room)[37:-6]
                power_review = str(power)[37:-6]
                control_review = str(control)[37:-6]
                fuel_review = str(fuel)[37:-6]
                comfort_review = str(comfort)[38:-6]
                outer_review = str(outer_view)[37:-6]
                inner_review = str(inner_view)[37:-6]
                cost_review = str(price)[38:-6]
            except:
                continue

            data.append([customer_name,present_time,car_type,purchase_time,purchase_place,car_price,average_mpg,drive_distance,overall_score,room_score,power_score,control_score,mpg_score,comfort_score,outer_score,cost_score,adv_review,dis_review,room_review,power_review,control_review,fuel_review,comfort_review,outer_review,inner_review,cost_review])
            
            

with open(path+'.csv', 'w', encoding='utf-16') as f:
    for text in data:
        f.write(text[0] + '\t' + text[1]+ '\t' + text[2] + '\t' + text[3] + '\t' + text[4]+'\t'+ text[5]+ '\t' + text[6] + '\t'+ text[7]+'\t' + text[8] + '\t' + text[9] + '\t' + text[10]+'\t'+ text[11]+ '\t' + text[12] + '\t'+ text[13] + '\t' + text[14] + '\t' + text[15]+ '\t' + text[16] + '\t' + text[17] + '\t' + text[18]+'\t'+ text[19]+ '\t' + text[20] + '\t'+ text[21]+'\t' + text[22] + '\t' + text[23] + '\t' + text[24] + '\t' + text[25] + '\n')            
    
    
            
                
            
                
```
