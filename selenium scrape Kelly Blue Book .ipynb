Using Selenium to scrape customer reviews from Kelly Blue Book top25 best selling suv in 2019

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import random
import re
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
import time
import json
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import os
import pandas as pd
```

```python
cartype = ['buick/encore/2019', 'volkswagen/tiguan/2019', 'lexus/rx/2019', 'chevrolet/trax/2019', 'hyundai/santa-fe/2019', 'subaru/crosstrek/2019', 'toyota/4runner/2019', 'honda/pilot/2019', 'hyundai/tucson/2019', 'ford/edge/2019', 'jeep/compass/2019', 'chevrolet/traverse/2019', 'mazda/cx-5/2019', 'subaru/forester/2019', 'subaru/outback/2019', 'ford/explorer/2019', 'jeep/cherokee/2019', 'jeep/wrangler/2019', 'toyota/highlander/2019', 'ford/escape/2019', 'jeep/grand-cherokee/2019', 'chevrolet/equinox/2019', 'nissan/rogue/2019', 'honda/cr-v/2019', 'toyota/rav4/2019']
len(cartype)
```

```python
all_reviews = []
all_years = []
all_car_types = []

#1.open the websites

driver = Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")

driver.get('https://www.kbb.com/' + carype + '/consumer-reviews/')
soup = BeautifulSoup(driver.page_source, 'lxml')
button = 'More Reviews'

#2.find the more reviews button and click

while button == 'More Reviews':
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    try:
        button = soup.find('button', {'data-analytics':re.compile('morereviews_btn')})
        button = button.text.strip()
        driver.find_element(By.CSS_SELECTOR, ".css-dma2fx-StyledButton-buttonStyles-secondary-default").click()
        time.sleep(1)
    except:
        button = 'none'

soup = BeautifulSoup(driver.page_source, 'lxml')

#3.extract review
review_l = []

for div_class in ['css-cumw5s','css-b7i47v']:

    contents = soup.find_all('div',{'class':div_class})

    for single_review in contents:
        for i in single_review:   
            target = str(i.find_all('span'))
            if len(target) != 2 and target[6] == '>':
                list1 = target.split('>, <')
                review_part = list1[0][7:]

                if review_part[-1] == ']':
                    review = review_part[:-8].replace('\n','')
                    review_l.append(review)
                else:
                    review = review_part[:-6].replace('\n','')
                    review_l.append(review)

#4.extract information of review
info_l_ori = []
year_l = []
car_type_l = []

for div_class in ['css-18pmvx8','css-1i2gasy']:

    contents = soup.find_all('div',{'class': div_class})

    for i in contents:
        info = i.text.strip()
        info_l_ori.append(info)


    for i in info_l_ori[0:len(review_l)]:
        slash1 = int(i.index('/'))
        year = i[slash1 + 4:slash1 + 8]
        car_type = i[slash1 + 8:]

        year_l.append(year)
        car_type_l.append(car_type)


5.#append reviews and informations of one ranked car in all_reviews and all_information
for i in range(0,len(review_l)):
    all_reviews.append(review_l[i])
    all_years.append(year_l[i])
    all_car_types.append(car_type_l[i])
driver.quit()
```

```python
df = pd.DataFrame(columns = ["Year", "Car Type", "Review"])
df['Year'] = all_years
df['Car Type'] = all_car_types
df['Review'] = all_reviews

df.head(100)
```
