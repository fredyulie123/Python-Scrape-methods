```python
import re
import requests
from bs4 import BeautifulSoup
import time
import os
import pandas as pd
import numpy as np
```

```python
# define notebook path
os.getcwd()
path = 'new'
```

```python
data = []

# set chrome environment
my_headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36', 'Accept-Language':'zh-CN'}
AUTOTHROTTLE_ENABLED = True

page = 6

for k in range(0, page+1):
    if k == 0 :
        url = "https://k.autohome.com.cn/5365/#dataList" # get urls
    else :
        url = "https://k.autohome.com.cn/5365/index_" +str(k)+ ".html#dataList"
    src = False
    for i in range(1,6):
        try:
            response = requests.get(url, headers = my_headers)
            src = response.content
            break
        except:
            print('failed attempt #', i)
            time.sleep(2)
    if not src:
        print('Could not get page', url)
    else:
        print('Successfully get the page', url)

# extract information
    soup = BeautifulSoup(src.decode('gb2312', 'ignore'),'lxml')
    
    infos = soup.findAll('div', {'class':re.compile('mouthcon')})
    
    for info in infos:
        k = info.findAll('div',{'class':re.compile('position-r')})
        m = info.findAll('div',{'class':re.compile('mouthcon-cont fn-clear')})
        for i in k:
            kj = k[0].find('span', {'class':re.compile('font-arial c333')})
            dl = k[1].find('span', {'class':re.compile('font-arial c333')})
            ck = k[2].find('span', {'class':re.compile('font-arial c333')})
            yh = k[3].find('span', {'class':re.compile('font-arial c333')})
            ssx = k[4].find('span', {'class':re.compile('font-arial c333')})
            wg = k[5].find('span', {'class':re.compile('font-arial c333')})
            ns = k[6].find('span', {'class':re.compile('font-arial c333')})
            xjb = k[7].find('span', {'class':re.compile('font-arial c333')})
        for c in m:
            ctype = info.find('span', {'class':re.compile('font-arial')})
            place = info.find('dd', {'class':re.compile('c333')})
            time = info.find('dd', {'class':re.compile('font-arial bg-blue')})
            reason = info.find('dd', {'class':re.compile('fn-clear')})
            sup = info.find('label', {'class':re.compile('supportNumber')})
            view = info.find('a', {'class':re.compile('orange'),'target':re.compile('_blank')})

            if ctype:
                car_type = ctype.text.strip()
            if place:
                p_place = place.text.strip()
            if time:
                p_time = time.text.strip()
            if sup:
                num_sup = sup.text.strip()
            if reason:
                m= reason.text.strip()
                r= m.split()
                p_reason = ' '.join(r)         
            if view:
                num_view = view.text.strip()
            if kj:
                kjn = kj.text.strip()
            if dl:
                dln = dl.text.strip()
            if ck:
                ckn = ck.text.strip()
            if yh:
                yhn = yh.text.strip()
            if ssx:
                ssxn = ssx.text.strip()
            if wg:
                wgn = wg.text.strip()
            if ns:
                nsn = ns.text.strip()
            if xjb:
                xjbn = xjb.text.strip()
            data.append([car_type, p_place, p_time, num_sup, num_view, p_reason, kjn, dln, ckn, yhn, ssxn, wgn, nsn, xjbn])

with open(path+'.txt', 'w', encoding='utf-8') as f:
    for text in data:
        f.write(text[0] + '\t' + text[1]+ '\t' + text[2] + '\t' + text[3] + '\t' + text[4]+'\t'+ text[5]+ '\t' + text[6] + '\t'+ text[7]+'\t' + text[8] + '\t' + text[9] + '\t' + text[10]+'\t'+ text[11]+ '\t' + text[12] + '\t'+ text[13]+ '\n')
```

```python
# add headers
user_cols=['车型','购车地点','购车时间','评分采纳量','浏览量','购车原因','空间','动力','操控','油耗','舒适性','外观','内饰','性价比']
users=pd.read_csv(path+'.txt',sep='\t', header=None, names=user_cols)
users.head()
```
